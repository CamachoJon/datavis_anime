```python
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt
import plotly.figure_factory as ff
import plotly.graph_objs as go
```

## Revenue data

```python
df = pd.read_csv('anime_revenue.csv')
```

```python
df.columns
```

```python
df.nunique().sort_values()
```

### Hist

```python
feature_to_plot = 'Revenue (Millions)'
feature_to_split = 'source' 

fig = px.histogram(df, x=feature_to_plot, color=feature_to_split, nbins=50, histnorm='probability density', 
                   color_discrete_sequence=px.colors.sequential.Agsunset)

fig.update_layout(
    title=f'Histogram of {feature_to_plot} splitting by {feature_to_split}',
    xaxis_title=feature_to_plot,
    yaxis_title=feature_to_split,
)

fig.show()
```

### Barplot

```python
feature = 'Revenue (Millions)'
top10_df = df.sort_values(feature, ascending=False).head(10)

fig = px.bar(
    top10_df,
    x='title',
    y=feature,
    text=feature,
    # color='Genre',
    hover_data=[feature],
    color_continuous_scale=px.colors.sequential.Agsunset
    )

fig.update_layout(
    title=f'Top 10 Animes by {feature}',
    xaxis_title='Anime Title',
    yaxis_title=feature,
    yaxis_tickformat='$,.2f'
)

fig.show()
```

## All data

```python
df = pd.read_csv('anime_filtered.csv')
df.head()
```

```python
df.columns
```

### Mosaic

```python
x_var = 'status'
y_var = 'source'

grouped_df = df.groupby([y_var, x_var]).size().reset_index(name='Count')

fig = px.treemap(
    grouped_df,
    path=[y_var, x_var],
    values='Count',
    color='Count',
    color_continuous_scale=px.colors.sequential.Agsunset)


fig.update_layout(
    title=f'Treemap of {x_var} and {y_var}',
)

# Show the plot
fig.show()
```

### Scatter

```python
x_var = 'members'
y_var = 'score' 
color = 'type'

fig = px.scatter(
    df,
    x=x_var,
    y=y_var,
    color=color,
    hover_data=['title'],
    opacity=0.7,
    color_discrete_sequence=px.colors.sequential.Agsunset
)

fig.update_layout(
    title=f'Scatter of {y_var} and {x_var}',
    xaxis_title=x_var,
    yaxis_title=y_var,
)

fig.show()
```

### Time

```python
df['Year'] = df['premiered'].str.extract('(\d{4})')
df['Season'] = df['premiered'].str.extract('(Winter|Spring|Summer|Fall)')

premieres = df.groupby(['Year', 'Season']).size().reset_index(name='Count')

fig = go.Figure()
seasons = ['Winter', 'Spring', 'Summer', 'Fall']
for season in seasons:
    season_data = premieres[premieres['Season'] == season]
    fig.add_trace(go.Scatter(
        x=season_data['Year'],
        y=season_data['Count'],
        mode='lines+markers',
        name=season,
        line=dict(
            color=px.colors.sequential.Agsunset[seasons.index(season)],
            width=3
        )
    ))

fig.update_layout(
    title='Number of Animes Premiered by Season',
    xaxis_title='Year',
    yaxis_title='Number of Animes',
    yaxis_tickformat=',d'
)

fig.show()
```
